##Plot example:

```python
image = nl.image.load_img('') #fill path to image
mask = nl.image.load_img('') #fill path to mask

fig, axes = plt.subplots(nrows=4, figsize=(30, 40))

nlplt.plot_anat(image,
                title='Anatomical view',
                axes=axes[0])

nlplt.plot_epi(image,
               title='Epi-view',
               axes=axes[1])

nlplt.plot_img(image,
               title='normal color view',
               axes=axes[2])

nlplt.plot_roi(mask, 
               title='Superposition with mask',
               bg_img=image, 
               axes=axes[3], cmap='Paired')

plt.show()
```
